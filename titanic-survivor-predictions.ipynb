```python
train_data['Sex'] = train_data['Sex'].astype("category")
train_data['Ticket'] = train_data['Ticket'].astype("category")
train_data['Embarked'] = train_data['Embarked'].astype("category")
train_data['Age'] = train_data['Age'].astype("int")
```

```python
test_data.drop(['SibSp','Parch','Cabin','Name'], inplace = True , axis=1)
```

```python
test_data['Sex'] = train_data['Sex'].astype("category")
test_data['Ticket'] = train_data['Ticket'].astype("category")
test_data['Embarked'] = train_data['Embarked'].astype("category")
test_data['Age'] = train_data['Age'].astype("int")
```

```python
test_data.info()
```

```python
bins = [0,18 ,30 ,45]
labels = ['Teen','Young','Adult']
train_data['AgeGroup'] = pd.cut(train_data['Age'], bins=bins, labels=labels, right=False)
test_data['AgeGroup'] = pd.cut(test_data['Age'], bins=bins, labels=labels, right=False)
```

```python
bins = [0,1 ,2 ,5,10]
labels = ['Single','duo','Small','Large']
test_data['Family_size'] = pd.cut(test_data['temp_family'], bins=bins, labels=labels, right=False)
train_data['Family_size'] = pd.cut(train_data['temp_family'], bins=bins, labels=labels, right=False)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
train_data['Sex'] = le.fit_transform(train_data['Sex'])
```

```python
test_data['Sex'] = le.fit_transform(test_data['Sex'])
```

```python
train_data['Embarked'] = le.fit_transform(train_data['Embarked'])
test_data['Embarked'] = le.fit_transform(test_data['Embarked'])
```

```python
train_data['AgeGroup'] = le.fit_transform(train_data['AgeGroup'])
train_data['Family_size'] = le.fit_transform(train_data['Family_size'])
test_data['AgeGroup'] = le.fit_transform(test_data['AgeGroup'])
test_data['Family_size'] = le.fit_transform(test_data['Family_size'])
```

```python
train_data.head()
```

```python
train_data.drop(['Age','Ticket'],inplace=True,axis=1)
test_data.drop(['Age','Ticket'],inplace=True,axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = train_data['Survived']
train_data.drop(['Survived'],inplace = True , axis=1)
```

```python
x = train_data
```

```python
x
```

```python
y
```

```python
x_train , x_test ,y_train , y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.tree import DecisionTreeRegressor
dc = DecisionTreeRegressor(random_state = 72)
dc.fit(x_train,y_train)
```

```python
y_pred = dc.predict(x_test)
accuracy_score(y_test , y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred , y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
accuracy_score(y_test , y_pred)
```

```python
print(classification_report(y_pred , y_test))
```

```python
pred = classifier.predict(test_data)
```
